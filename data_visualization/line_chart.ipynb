# Basic line chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create some random data
x = np.linspace(start=0, stop=10, num=100)
y = np.sin(x)
```

```python
# Create a line chart of the data
plt.plot(x, y)
# Add labels and a title to the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Chart Example')
# Show the plot
plt.show()
```

## Line chart with multiple data sets and legend

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create some random data
x = np.linspace(start=0, stop=10, num=100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# Create a line chart of the data with a legend
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
# Add a legend, labels, and a title to the plot
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Chart Example with Legend')
# Show the plot
plt.show()
```

# linestyle, marker, and color paramters

Here are some characters that can be used for `linestyle` or `marker` parameters.  


| character | description           |   
|:-----------|:-----------------------|
| '-'       | solid line style      |
| '--'      | dashed line style     |
| '-.'      | dash-dot line style   |
| ':'       | dotted line style     |
| '.'       | point marker          |
| ','       | pixel marker          |
| 'o'       | circle marker         |
| 'v'       | triangle_down marker  |
| '^'       | triangle_up marker    |
| '<'       | triangle_left marker  |
| '>'       | triangle_right marker |
| '1'       | tri_down marker       |
| '2'       | tri_up marker         |
| '3'       | tri_left marker       |
| '4'       | tri_right marker      |
| 's'       | square marker         |
| 'p'       | pentagon marker       |


Below are characters that can be used for `color` parameter.

| character | color                 |
|:-----------|:-----------------------|
| b       | blue                  |
| g       | green                 |
| r       | red                   |
| c       | cyan                  |
| m       | magenta               |
| y       | yellow                |
| k       | black                 |
| w       | white                 |


For more styles, refer to [this](https://matplotlib.org/2.1.2/api/_as_gen/matplotlib.pyplot.plot.html) page.   


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create some random data
x = np.linspace(start=0, stop=10, num=100)
y = np.sin(x)
```

```python
plt.plot(x, y, linestyle="--", color='m')
plt.show()
```

```python
plt.plot(x, y, marker="v", color='c')
plt.show()
```

# Plotting 3D data (plot3D)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate some random 3D data
x = np.linspace(0, 20, 100)
y = np.sin(x)
z = x * y
```

```python
fig = plt.figure()
# Create a subplot
ax = fig.add_subplot(111, projection="3d")
# Plot data
ax.plot3D(xs=x, ys=y, zs=z, c='gray')
# Set labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Plot Example')
plt.show()
```

# Plotting 3D data (plot_surface)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate some data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
X.shape, Y.shape
```

```python
plt.scatter(X, Y, s=1)
plt.show()
```

```python
# Create Z1, Z2 data
Z1 = np.sin(np.sqrt(X**2 + Y**2))
Z2 = np.cos(np.sqrt(X**2 + Y**2))
```

```python
# Create a new figure object
fig = plt.figure(figsize=(8, 4))

# Add the first subplot
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z1, cmap='viridis')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')

# Add the second subplot
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X, Y, Z2, cmap='plasma')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')

# Show the plot
plt.show()
```

# Plotting 3D data (wire_frame)

```python
# Create a new figure object
fig = plt.figure()

# Add a subplot
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z1, color='green', alpha=0.5)
ax.plot_wireframe(X, Y, Z2, color='yellow', alpha=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("wireframe")

# Show the plot
plt.show()
```
