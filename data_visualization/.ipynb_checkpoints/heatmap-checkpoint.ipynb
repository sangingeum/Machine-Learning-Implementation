```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Basic Heatmap

```python
data = np.random.rand(10, 10)
data.shape
```

```python
sns.heatmap(data)
```

# Heatmap with Annotations
`annot`: bool or rectangular data    
If it is True, write the data value in the cell.   
If it is rectangular data which has the same shape as data, the use this to annotate the heatmap instead of data.   
`fmt`: string formatting code

```python
sns.heatmap(data, annot=True, fmt='.2f')
```

# cmap, center, linewidths, and linecolor parameters
`cmap`: The mapping from data values to color space. Matplotlib colormaps can be used.    
`center`: The value at which to center the colormap.   
`linewidth`: Width of the lines that will divide each cell.    
`linecolor`: Color of the lines that will divide each cell.

```python
sns.heatmap(data, annot=True, cmap=plt.cm.Blues, center=0.2, linewidth=2, linecolor='skyblue', fmt='.2f')
```

# xticklabels, yticklabels, cbar parameters
`xticklabels`, `yticklabels`:     
If True, plot the column names of the dataframe.    
If False, don't plot the column names.     
If list-like, plot element of the list as xticklabels.     
`cbar`: Whether to draw a color bar. Boolean.   


```python
xlabels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
ylabels = ['aa', 'bb', 'cc', 'dd', 'ee', 'ff', 'gg', 'hh', 'ii', 'jj']
```

```python
sns.heatmap(data, annot=True, cmap=plt.cm.Greens, fmt='.2f', cbar=False, xticklabels=xlabels, yticklabels=ylabels)
```

# Robust parameter
`robust`: If True and vmin or vmax are absent, the colormap range is computed with robust quantiles instead of extreme values.

```python
# insert outliers
data[0][0] = 100
data[0][1] = -100
```

```python
# When robust = False
sns.heatmap(data, annot=True, cmap=plt.cm.Oranges, fmt='.2f')
```

```python
# When robust = True
sns.heatmap(data, annot=True, cmap=plt.cm.Oranges, fmt='.2f', robust=True)
```
