# Simple Bar Chart
A simple bar chart with a legend, labels for x and y axes, and a title.

```python
import matplotlib.pyplot as plt
```

```python
# data
x = ['A', 'B', 'C', 'D', 'E']
y = [25, 40, 30, 35, 20]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
# draw a bar chart
ax.bar(x, y, label='data 1')
# Add a legend
ax.legend()
# Set the x-axis label
ax.set_xlabel('Category')
# Set the y-axis label
ax.set_ylabel('Value')
# Set the title of the plot
ax.set_title('Bar Chart Example')
# Display the plot
plt.show()
```

# Horizontal Bar Chart

We create a horizontal bar chart using the `barh()` function instead of the `bar()` function.     
Note that x and y labels are swapped.

```python
import matplotlib.pyplot as plt
```

```python
# Data
x = ['A', 'B', 'C', 'D', 'E']
y = [25, 40, 30, 35, 20]
```

```python
# Create a figure and a plot
fig = plt.figure()
ax = fig.add_subplot(111)
# horizontal bar chart
ax.barh(x, y)

ax.set_xlabel("value")
ax.set_ylabel("category")
ax.set_title("Horizontal Bar Chart Example")
plt.show()
```

# Bar Chart with Multiple Categories
We can plot multiple datasets by adding or subtracting offsets to the x-coordinates of the default positions; `np.arange(len(categories))`.    
We use the `set_xticks()` method to set the category values.


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# data
categories = ['A', 'B', 'C', 'D', 'E']
values1 = [25, 40, 30, 35, 20]
values2 = [30, 20, 45, 25, 15]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

bar_width = 0.35

# Make bars 
x = np.arange(len(categories))
ax.bar(x - bar_width/2, values1, bar_width, label='Dataset 1', capsize=5)
ax.bar(x + bar_width/2, values2, bar_width, label='Dataset 2', capsize=5)

# Apply category values
ax.set_xticks(x, categories)
# Add a legend
ax.legend()
# Set the x-axis label
ax.set_xlabel('Category')
# Set the y-axis label
ax.set_ylabel('Value')
# Set the title of the plot
ax.set_title('Vertical Bar Chart Example')
# Show the plot
plt.show()
```

# Bar chart with errors
We can draw error lines by using the `yerr` or `xerr` parameters.       
The `capsize` parameter determines the length of the caps around the error lines.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
errors1 = [2, 4, 3, 1, 5]
errors2 = [3, 1, 4, 2, 3]
```

```python
fig, ax = plt.subplots()
bar_width = 0.35
x = np.arange(len(categories))

# Set yerr, capsize parameters
ax.bar(x - bar_width/2, values1, bar_width, label='Dataset 1', yerr=errors1, capsize=5)
ax.bar(x + bar_width/2, values2, bar_width, label='Dataset 2', yerr=errors2, capsize=5)

ax.set_xticks(x, categories)
ax.set_xlabel('Category')
ax.set_ylabel('Value')
ax.set_title('Bar chart with errors')
ax.legend()
plt.show()
```

# Bar Chart with Data Labels and Limits
We can specify data labels by calling the `bar_label()` method.   
The upper and lower limits of y axis can be set by using the `set_ylim()` method. `set_xlim()` method also exists.

```python
fig, ax = plt.subplots()
bar_width = 0.35
x = np.arange(len(categories))
rects1 = ax.bar(x - bar_width/2, values1, bar_width, label='Dataset 1', yerr=errors1, capsize=5)
rects2 = ax.bar(x + bar_width/2, values2, bar_width, label='Dataset 2', yerr=errors2, capsize=5)

# Set data labels
ax.bar_label(rects1, values1)
ax.bar_label(rects2, values2)

# Set the y limits.
ax.set_ylim(bottom=5, top=70)

ax.set_xticks(x, categories)
ax.set_xlabel('Category')
ax.set_ylabel('Value')
ax.set_title('Bar Chart with Data Labels and Limits')
ax.legend()
plt.show()
```

# Visualization of Classification Data

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000, n_informative=5, n_classes=5)
X.shape, y.shape
```

```python
# Get distribution of y
unique, counts = np.unique(y, return_counts=True)
unique, counts 
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
rects = ax.bar(unique, counts, color='green')
ax.bar_label(rects, counts)
ax.set_ylim(170, 230)
ax.set_xlabel('Class')
ax.set_ylabel('Count')
ax.set_title('Distribution of Y')
plt.show()
```
