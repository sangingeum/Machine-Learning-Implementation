# Simple Histogram
`bins` parameter takes a seqeunce of integers, an integer, or a string.    
If `bins` is a sequence,it defines the bin edges.     
If `bins` is an integer, it defines the number of equal-width bins in the range.     
If `bins` is a string, it is one of the binning strategies supported by `numpy.histogram_bin_edges`: 'auto', 'fd', 'doane', 'scott', 'stone', 'rice', 'sturges', or 'sqrt'.



````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
data = [87, 53, 66, 61, 67, 68, 62,
          110, 104, 61, 111, 123, 117,
          119, 116, 104, 92, 111, 90,
          103, 81, 80, 101, 51, 79, 107,
          110, 129, 145, 139, 110]
````

````python
# Make a histogram with 10 equal-width bins
plt.hist(data, bins=10, edgecolor="white", color="gray")
# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
# Show plot
plt.show()
````

````python
# auto
plt.hist(data, bins='auto', edgecolor="white", color="gray")
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
````

````python
np.min(data), np.max(data)
````

````python
# Manually setting bin edges
# Note that there are 4 bars when len(bins) is 5.
plt.hist(data, bins=[50, 75, 100, 125, 150], edgecolor="white", color="gray")
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
````

# Cummulative Histogram
You can make a cummulative histogram by setting the `cumulative` paramter to ```True```

````python
plt.hist(data, bins=[50, 75, 100, 125, 150], edgecolor="white", color="gray", cumulative=True)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
````

# Histogram with Multiple Data
### Unstacked Histogram
If `stack` is `False`, multiple data are arranged side by side.

````python
# Create two datasets
data1 = np.random.binomial(n=100, p=0.2, size=50).reshape((-1, 1))
data2 = np.random.binomial(n=100, p=0.3, size=50).reshape((-1, 1))
data = np.hstack([data1, data2])
np.min(data1), np.max(data1), np.min(data2), np.max(data2), data.shape
````

````python
# Create bins
bins = np.linspace(start=0, stop=50, num=6)
bins
````

````python
plt.hist(data, bins=bins, color=['orange', 'green'], edgecolor='w', label=['p=0.2', 'p=0.3'])
plt.xlabel("value")
plt.ylabel("count")
plt.title('Stacked Histogram')
plt.legend()
plt.show()
````

### Stacked Histogram
If `stacked` is `True`, multiple data are stacked on top of each other.    



````python
plt.hist(data, bins=bins, color=['orange', 'green'], edgecolor='w', stacked=True, label=['p=0.2', 'p=0.3'])
plt.xlabel("value")
plt.ylabel("count")
plt.title('Stacked Histogram')
plt.legend()
plt.show()
````

### Step Histogram
`histtype='step', fill=False`

````python
plt.hist(data, bins=bins, histtype='step', color=['orange', 'green'], stacked=True, label=['p=0.2', 'p=0.3'], fill=False)
plt.xlabel("value")
plt.ylabel("count")
plt.title('Step Histogram')
plt.legend()
plt.show()
````
