# 2D scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate some random data
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
# Create a 2D scatter plot
plt.scatter(x=x, y=y)
# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")
plt.show()
```

# c, cmap parameters
`c` parameter is used to specify a list of colors for each point in the plot.    
`cmap` parameter is used to specify a colormap for the colors.    
Predfined colormaps are avalable in `plt.cm`.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate some random data
x = np.random.rand(50)
y = np.random.rand(50)
c = np.round(np.random.rand(50))
np.unique(c, return_counts=True)
```

```python
# Create a scatter plot
plt.scatter(x=x, y=y, c=c, cmap=plt.cm.cool)
# Add a color bar to the plot
plt.colorbar()
# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")
plt.show()
```

# alpha, s parameters
`alpha` parameter is used to adjust the transparency of each point.   
`s` parameter is used to specify a list of sizes for each point.

```python
alpha = np.random.uniform(low=0.1, high=1.0, size=50)
s = np.random.randint(low=30, high=200, size=50)
np.min(alpha), np.max(alpha), np.min(s), np.max(s)
```

```python
# Create a scatter plot
plt.scatter(x=x, y=y, c=c, cmap=plt.cm.cool, s=s, alpha=alpha)
# Add a color bar to the plot
plt.colorbar()
# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")
plt.show()
```

# 3D scatter plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate some random data
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
c = np.round(np.random.rand(50))
s = np.random.randint(low=10, high=100, size=50)
alpha = np.random.uniform(low=0.1, high=1.0, size=50)
```

```python
# Create a 3D scatter plot
fig = plt.figure() # create a new figure
ax = fig.add_subplot(111, projection='3d') # Add a new set of axes to the figure
# '111' => a three-digit integer that represents the number of rows, columns, and position of the subplot.
ax.scatter(x, y, z, c=c, s=s, alpha=alpha, cmap=plt.cm.Oranges)

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Scatter Plot Example')

# Show the plot
plt.show()
```

## Visualization of circle data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
```

```python
X, y = make_circles(n_samples=1000, shuffle=True, noise=0.05)
X.shape, y.shape, np.unique(y, return_counts=True)
```

```python
plt.scatter(x=X[:,0], y=X[:,1], c=y, cmap=plt.cm.RdYlGn)
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Circle data")
plt.show()
```
